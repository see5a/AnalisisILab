```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
años = 5
dt = 1/12  #Paso mensual
n_pasos = int(años / dt)

#Datos iniciales
I0 = 40000  #Casos acumulados aproximados en 2020
r_anual = 1100  #Casos nuevos promedio por año
r_mensual = r_anual / 12

# Modelo: dI/dt = r(t)
# Para simplicidad, asumimos r constante; se puede modificar para r(t) variable

# Vector para almacenar resultados
I = np.zeros(n_pasos+1)
t = np.linspace(0, años, n_pasos+1)
I[0] = I0

# Método de Euler
for i in range(n_pasos):
    dI_dt = r_mensual  # tasa constante de nuevos casos por mes
    I[i+1] = I[i] + dI_dt * dt

# Graficar resultados
plt.figure(figsize=(10,6))
plt.plot(t + 2020, I, label='Casos acumulados de VIH', color='purple')
plt.xlabel('Año')
plt.ylabel('Número acumulado de casos')
plt.title('Incremento de casos acumulados de VIH en Honduras (2020-2025)')
plt.grid(True)
plt.legend()
plt.show()
```
